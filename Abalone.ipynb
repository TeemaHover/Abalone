# Type of Feature

The features of given data frame are quantitative (numeric). Let’s see mode of features to define their type.

```python
dat = abalone

def Mode(x):
    a = pd.Series(x).value_counts()
    return a.idxmax()

mode_result = Mode(dat['Sex'])

print("Mode: ", mode_result)
```

```python
mode_result = Mode(dat['Length'])

print("Mode: ", mode_result)
```

```python
mode_result = Mode(dat['Diameter'])

print("Mode: ", mode_result)
```

```python
mode_result = Mode(dat['Height'])

print("Mode: ", mode_result)
```

```python
mode_result = Mode(dat['Whole weight'])

print("Mode: ", mode_result)
```

```python
mode_result = Mode(dat['Shucked weight'])

print("Mode: ", mode_result)
```

```python
mode_result = Mode(dat['Viscera weight'])

print("Mode: ", mode_result)
```

```python
mode_result = Mode(dat['Shell weight'])

print("Mode: ", mode_result)
```

```python
mode_result = Mode(dat['Rings'])

print("Mode: ", mode_result)
```
