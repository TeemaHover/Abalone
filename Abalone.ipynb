# Type of Feature

The features of given data frame are quantitative (numeric). Let’s see mode of features to define their type.

```python
dat = abalone

def Mode(x):
    a = pd.Series(x).value_counts()
    return a.idxmax()

mode_result = Mode(dat['Sex'])

print("Mode: ", mode_result)
```

# Two-Sample Kolmogorov-Smirnov Test

```python
from scipy.stats import ks_2samp

male_lengths = abalone[abalone['Sex'] == 'M']['Length']
female_lengths = abalone[abalone['Sex'] == 'F']['Length']

ks_statistic, p_value = ks_2samp(male_lengths, female_lengths)
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")

```

```python
male_diameter = abalone[abalone['Sex'] == 'M']['Diameter']
female_diameter = abalone[abalone['Sex'] == 'F']['Diameter']

ks_statistic, p_value = ks_2samp(male_diameter, female_diameter)
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")
```

```python
male_height = abalone[abalone['Sex'] == 'M']['Height']
female_height = abalone[abalone['Sex'] == 'F']['Height']

ks_statistic, p_value = ks_2samp(male_height, female_height)
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")
```

```python
male = abalone[abalone['Sex'] == 'M']['Whole weight']
female = abalone[abalone['Sex'] == 'F']['Whole weight']

ks_statistic, p_value = ks_2samp(male, female)
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")
```

```python
male = abalone[abalone['Sex'] == 'M']['Shucked weight']
female = abalone[abalone['Sex'] == 'F']['Shucked weight']

ks_statistic, p_value = ks_2samp(male, female)
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")
```

```python
male = abalone[abalone['Sex'] == 'M']['Viscera weight']
female = abalone[abalone['Sex'] == 'F']['Viscera weight']

ks_statistic, p_value = ks_2samp(male, female)
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")
```

```python
male = abalone[abalone['Sex'] == 'M']['Shell weight']
female = abalone[abalone['Sex'] == 'F']['Shell weight']

ks_statistic, p_value = ks_2samp(male, female)
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")
```

```python
male = abalone[abalone['Sex'] == 'M']['Rings']
female = abalone[abalone['Sex'] == 'F']['Rings']

ks_statistic, p_value = ks_2samp(male, female)
print(f"KS Statistic: {ks_statistic}")
print(f"P-value: {p_value}")
```

# Correlation plot

```python
plt.figure(figsize=(10,8), dpi= 80)

abalone_re = pd.read_csv("data/abalone.csv")

numeric_abalone = abalone_re.select_dtypes(include=['float64', 'int64'])


plt.figure(figsize=(10, 8))
sns.heatmap(numeric_abalone.corr(), cmap='RdYlGn', center=0, annot=True)
plt.title('Correlation Matrix of Numeric Features in Abalone Dataset')
plt.show()
```

# Calculate mean median standard deviation

```python
grouped_abalone = abalone_re.groupby('Sex')


mean_values = grouped_abalone.mean()
median_values = grouped_abalone.median()
std_values = grouped_abalone.std()


print("Mean:")
print(mean_values)
print("\nMedian:")
print(median_values)
print("\nStandard Deviation:")
print(std_values)
```
